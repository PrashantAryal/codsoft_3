## Exploratory Data Analysis (EDA)

```python
sns.pairplot(df, hue='species', diag_kind='kde', palette='Set2')
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors = ['blue', 'orange', 'green', 'red']

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes = axes.flatten()

for i, (feature, color) in enumerate(zip(features, colors)):
    ax = axes[i]
    for species in df['species'].unique():
        sns.histplot(df[df['species'] == species][feature], kde=True, label=species,alpha=0.6, ax=ax)
    ax.set_title(f"Distribution of {feature.capitalize()} by Species")
    ax.set_xlabel(feature.capitalize())
    ax.set_ylabel("Frequency")
    ax.legend(df['species'].unique())

plt.tight_layout()
plt.show()
```

```python
df.hist(bins=20, figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
means = df.groupby('species').mean().reset_index()
means_melted = means.melt(id_vars='species', var_name='Feature', value_name='Mean')

plt.figure(figsize=(12, 6))
sns.barplot(x='Feature', y='Mean', hue='species', data=means_melted, palette='Set2')
plt.title("Mean Feature Values by Species")
plt.xlabel("Features")
plt.ylabel("Mean Value")
plt.legend(title="Species")
plt.show()
```

## Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['species'] = labelencoder.fit_transform(df['species'])
```

# Splitting the data and Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

x = df.drop(columns=['species'])
y = df['species']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
def evaluate_model(model,model_name):
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)

    print(f"***************--{model_name}--***************")
    print("\nClassification Report: ")
    print(classification_report(y_test,y_pred,target_names = labelencoder.classes_))

    print("\nConfusion Matrix")
    cm = confusion_matrix(y_test,y_pred)
    sns.heatmap(cm,cmap='inferno',annot = True, fmt = 'd',xticklabels=labelencoder.classes_, yticklabels=labelencoder.classes_)
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.2f}")
    print("\n" + "-" * 50 + "\n")
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
evaluate_model(lr, "Logistic Regression")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
evaluate_model(dt, "Decision Tree")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42, n_estimators=100)
evaluate_model(rf, "Random Forest")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
evaluate_model(knn, "K-Nearest Neighbors")
```
